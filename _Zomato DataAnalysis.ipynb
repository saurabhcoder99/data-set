```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# step 2 - create the data frame

```python
dataframe = pd.read_csv("Zomato data 121.csv")
print(dataframe)
```

```python
dataframe
```

# convert the  data type of column - rate

```python
def handleRate(value):
    value = str(value).split("/")
    value = value[0];
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
dataframe.head()
```

```python
dataframe.info()
```

# type of resturent

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlable("types of resturant")
```

# conclusion: Majority of the resturant falls in dinning category 

# Objective 2> How many votes has each types of resturant recived from customers ?

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result , c = "green",marker = "o")
plt.xlabel("Types of Resturent", c = "red",size = 20)
plt.xlabel("votes", c = "blue" , size = 20)
```

# conclusion Dinning Resturant has received maximum votes

# Objective 3> what are the ratings that the majority of Resturants have received ?

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins = 5)
plt.title("rating distribution")
plt.show()
```

# Conclusion: the majority resturents received ratings frome 3.5 to 4

# Objective 6> Which type pf resturant received more offline oreders, so that Zomato can provide customers with some good offers ?

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index = 'listed_in(type)', columns = 'online_order', aggfunc = 'size', fill_value = 0)
sns.heatmap(pivot_table,annot = True, cmap="YlGnBu" , fmt = 'd')
plt.title('Heatmap')
plt.xlabel('Online_Order')
plt.ylabel("Listed In (Type)")
plt.show()
```

## Conclusion : Dinning Resturants primarily accept offline order
